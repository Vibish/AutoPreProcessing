```python
import scipy.stats as stats
for CategoricalVar in eda.CategoricalFeatures:
    Binary = 'Yes' if CategoricalVar in eda.BinaryCategoricalFeatures else 'No'
    for ContinuousVar in eda.ContinuousFeatures:
        f,p = stats.f_oneway(*[list(eda.df[eda.df[CategoricalVar]==name][ContinuousVar].dropna()) for name in set(eda.df[CategoricalVar].dropna())])
        print(f,p)
```
